## Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Line Plot

```python
# Generate 200 points between 0 and 10
x = np.linspace(0, 10, 100)
```

```python
10/200
```

```python
y = np.sin(x)
```

```python
plt.plot(x, y, 'red')
plt.xlabel('x')
plt.ylabel('Sin(x)')
plt.title('Graph of Sin(x)')
plt.show()
```

```python
y = np.cos(x)
```

```python
plt.plot(x, y, 'blue')
plt.xlabel('x')
plt.ylabel('Cos(x)')
plt.title('Graph of Cos(x)')
plt.show()
```

### Scatter Plot

```python
x = np.array([1,2,3,4,5,6,7])
y = np.array([10,5,7,6,11,6,7])
```

```python
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### Histograms

```python
y = np.random.random(1000)
```

```python
plt.hist(y, bins=20)
plt.ylabel('y')
plt.show()
```
